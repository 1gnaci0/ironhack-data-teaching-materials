## Boolean expressions

A boolean expression is an expression that that evaluates to either `True` or `False`. 

These two are a special "types" in python called `Bool`.  

In Python there exist the following [relational operators](https://www.geeksforgeeks.org/relational-operators-in-python), which represent boolean expressions:

*      `x` == `y`               # x is equal to y
*      `x` != `y`               # x is not equal to y
*      `x` > `y`                # x is greater than y
*      `x` < `y`                # x is less than y
*      `x` >= `y`               # x is greater than or equal to y
*      `x` <= `y`               # x is less than or equal to y
      

Note the difference between a single equal sign (=), which is an assignment, and a double equal sign (==), which is a relational operator

**Tip:** Use parenthesis to keep your operations in order!


```python
# Let´s Play with them!
x = 2
y = 4

print(y == (2 * x))
```

With solely these two values `True`|`False` we can create an entire branch of mathematics called [Boolean Algebra](https://en.wikipedia.org/wiki/Boolean_algebra#Laws). Whereas in regular Algebra de basic operations are addition and multiplication, the main operations of Boolean algebra are the conjunction (and) denoted as ∧, the disjunction (or) denoted as ∨, and the negation (not) denoted as ¬. **It is the formalism used for describing logical operations**.

In [Python we write this operations](https://www.geeksforgeeks.org/python-3-logical-operators/?ref=rp) as:

* `x` and `y` 
* `x` or `y`
* not `x`

Although the meaning of these operations is clear, we can completely define them with the so called "truth table":


```python
alice_is_rich = True
alice_lives_in_new_york = False
# Truth tables
    # True (operator) True -> ?
    # True (operator) False -> ?
    # False (operator) True -> ?
    # False (operator) False -> ? 
```

Things can get trickier when we combine several logical expressions... Consider for example the so-called [De Morgan´s laws](https://en.wikipedia.org/wiki/De_Morgan%27s_laws):

* not (A or B) = not A and not B
* not (A and B) = not A or not B

[Boolean algebra](https://en.wikipedia.org/wiki/Boolean_algebra#Laws) satisfy many common rules such as commutativity, associativity, etc. But we rather try to keep it simple here and learn by doing!

### else
Sometimes you want that if the condition is not satisfied (and only then) another action is executed. This is achieved by the `else` statement

```python
condition = False

if (condition):
    print('the condition is satisified')
else:
    print('the condition is not satisified')
```

### elif
Sometimes there are more than two possibilities. Think of something like:
"If Bob is allow to enter the disco, let him in. Else, if he is above 16 recommend him the light disco nearby. Else, send him home"

We can achieve this with the `elif` clause and chained conditionals:

```python
condition_1 = False
condition_2 = True

if (condition_2):
    print('the condition 1 is satisified')

elif (condition_2):
    print('the condition 2 is satisified (but not the first!)')
else:
    print('No condition is not satisified')
```

### pass
The pass statement does nothing. It can be used when a statement is required syntactically  but the program requires no action (from the [docs](https://docs.python.org/3/tutorial/controlflow.html))

```python
max_disco_capacity = 50 # Remember, this is 2020...
current_people_in = 10

while (current_people_in < max_disco_capacity):
    # let the newcomer enter!
    current_people_in = current_people_in + 1
    
print(current_people_in)
```

## The "for" statement

Python’s `for` statement iterates over the items of any sequence (a list or a string), in the order that they appear in the sequence. (from the [docs](https://docs.python.org/3/tutorial/controlflow.html))


```python
# Measure some strings:
words = ['cat', 'window', 'defenestrate']
for w in words:
    print(w, len(w))
```

### break
We can literally `break` a recursion when a condition is satisfied

```python
# Let stop the loop when we find the word 'window'
words = ['cat', 'window', 'defenestrate']
for w in words:
    print(w, len(w))
    if w == 'window':
        break
```

### range()

If you do need to iterate over a sequence of numbers, the built-in function `range()` comes in handy. It generates arithmetic progressions:

```python
for i in range(5):
    print(i)
```

### enumerate()
Similarly `enumerate` is widely used:

```python
list_of_names =['Alice', 'Bob', 'Jack', 'Mary', 'Manuel', 'Amanda', 'Ignacio', 'Fernando']
for i, name in enumerate(list_of_names):
    print(i, name)
```

### continue
The `continue` statement, continues with the next iteration of the loop, so that everything after it is skipped:


```python
# Program to show the use of continue statement inside loops
name_not_to_print = 'Ignacio'
for name in list_of_names:
    if name == name_not_to_print:
        continue
    # when the condition is satisfied, then continues prevents the print statement
    print(name)
```
